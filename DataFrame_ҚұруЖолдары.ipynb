## Сөздіктің көмегімен құру

## Series объектісінен тұратын тізімнің көмегімен құру

```python
temps_at_time0 = pd.Series([70, 71])
temps_at_time1 = pd.Series([90, 91])

df = pd.DataFrame([temps_at_time0, temps_at_time1])
```

```python
df
```

```python
df = pd.DataFrame([temps_at_time0, temps_at_time1], columns = ['Missoula', 'Philadelphia'])
```

```python
df
```

```python
df = pd.DataFrame([temps_at_time0, temps_at_time1])
df.columns = ['Missoula', 'Philadelphia']
```

```python
df
```

```python
pd.DataFrame({'Missoula':temps_at_time0, 'Philadelphia':temps_at_time1})
```

```python
temps_at_time2 = pd.Series([85, 87], index = [1,2])
df = pd.DataFrame({
        'Missoula': temps_at_time0,
        'Philadelphia': temps_at_time1,
        'New York': temps_at_time2
    })
df
```
